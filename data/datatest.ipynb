```python
import pandas as pd

# Load the dataset
df = pd.read_csv("Water_data.csv")
```

```python
print("="*60)
print("DATASET INFORMATION")
print("="*60)
```

```python
print(f"\nDataset shape: {df.shape}")
print(f"Total rows: {len(df)}")

print("\nColumn names and types:")
print(df.dtypes)

print("\n" + "="*60)
print("CATEGORY ANALYSIS")
print("="*60)

if 'category' in df.columns:
    print("\nUnique categories in your dataset:")
    categories = df['category'].unique()
    for cat in categories:
        print(f"  - {cat}")
    
    print("\nCategory distribution:")
    print(df['category'].value_counts())
    
    print("\n" + "="*60)
    print("SAMPLE DATA FOR EACH CATEGORY")
    print("="*60)
    
    for cat in categories:
        print(f"\n{cat} - Sample data:")
        print(df[df['category'] == cat].head(3))
        
else:
    print("\n⚠️ No 'category' column found!")

print("\n" + "="*60)
print("FIRST 10 ROWS OF DATASET")
print("="*60)
print(df.head(10))
```
